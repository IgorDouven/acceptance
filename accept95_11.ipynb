# Acceptance rules for Bayesians

### Introduction

Acceptance rules are associated with classical (frequentist) statistics (e.g., accept the 'alternative' hypothesis&mdash;typically the hypothesis of interest&mdash;if $p<.05$). But several authors have argued that Bayesians need acceptance rules as well: the statistics yields probabilities, but we are frequently in a situation in which we have to make a categorical claim with confidence. For example, companies may have to make categorical claims to reach a targeted group of consumers, given that probabilistic information would generally be lost on them. We expect such categorical claims to be connected to the probabilities we got from the statistics, and acceptance rules are supposed to systematize the connection.

Once we have acceptance rules (whatever the underlying statistics), we can ask about various types of error. In classical statistics, the key error types are commonly referred to as 'Type I errors' and 'Type II errors', where the former designate the type of mistakes of rejecting the null hypothesis even though it is true while the latter are the type of mistakes of rejecting the alternative hypothesis while the null hypothesis is actually false. For any given acceptance rule, we can ask for either type of error what the likelihood is with which we will commit it if we assume that rule.

Bayesians have argued that, while the idea of Type I and Type II errors may make little sense, given that a null hypothesis is virtually *never* true (a requirement for making a Type I error) nor should we ever be tempted to *claim* that it is true (a requirement for making a Type II error). But they have identified errors in the vicinity of Type I and Type II errors that do make sense from a Bayesian perspective, such as Type S errors&mdash;the error of wrongly claiming with confidence that a quantity has a given *sign* (e.g., that the effect of the treatment is positive)&mdash;and Type M errors&mdash;making wrong claims with confidence about the *magnitude* of some measured effect (e.g., claiming that a new treatment is *highly* effective when in reality it only has a small effect); see, for instance [this](http://www.stat.columbia.edu/~gelman/research/published/francis8.pdf) or [this](http://www.stat.columbia.edu/~gelman/research/published/retropower_final.pdf) paper.

Acceptance rules can fall short in other respects as well. We often do not just want to make claims with confidence; we want to be in a position to make such claims fast. For instance, a research team may want to announce that they have isolated a new type of virus as the cause of some disease (they do not want to accounce that their evidence warrants a confidence of so-and-so much percent that the virus causes the disease&mdash;that is much less likely to make headlines). It is important to them that they make the announcement *fast*, knowing that other teams are also trying to determine the cause of the given disease. On the other hand, they do not want to make a claim that they may later have to retract, to the embarrasment of the lab. For this reason, our acceptance rule should be (what we shall call) *cautious*, so as to avoid hasty announcements (or, more generally, hasty acceptance), but it should not be *overly* cautious, so as to avoid being typically slower in a position to announce or accept something than the competition. More generally, it should make the right tradeoff between speed and accuracy.

The acceptance rules we shall be interested in have the special feature that they are supposed to select *one* hypothesis from a partition of hypotheses of interest. To be more precise, let $\mathcal{H}=\{H_1,\ldots,H_n\}$ be a hypothesis partition, that is, a set of self-consistent, mutually exclusive, and jointly exhaustive hypotheses, where the members of $\mathcal{H}$ are the only propositions of interest in the present context. Then we will be looking at acceptance rules $\mathfrak{r}(\mathcal{H},\mathcal{E})=\mathtt{ans}$, where $\mathcal{E}$ is a body of evidence, and where either $\mathtt{ans}=\mathtt{void}$ or $\mathtt{ans}=i$, for some $i\colon 1\leqslant i\leqslant n$. The interpretation is that either the rule outputs the index of the member of $\mathcal{H}$ that is acceptable on the basis of $\mathcal{E}$ or it does not give any output, meaning that $\mathcal{E}$ is not sufficient evidence to accept any of the hypotheses.

Note in particular that acceptance rules do not yield as output disjunctions of hypotheses. In many situations, we want an acceptance rule exactlty of the kind here defined. For instance, a health insurer will want to be able to claim with confidence that a treatment is effective before it decides to pay for the treatment. That the treatment is either effective or that it makes no difference to the patient's status (so, e.g., just being able to exclude with confidence that it negatively affects the patient's health) will normally not be enough. Similarly, a doctor will want to be in a position where she can claim with confidence what *exactly* is wrong with a patient before she starts a treatment with possible long-term adverse consequences; for instance, she will want to be able to exclude with confidence that the patient is suffering from pneumonia or bronchitis or ... before she decides to start chemotherapy for lung cancer. These are the kinds of situation where we want to able to accept exactly one member of a hypothesis partition, rather than a disjunction of members.

Formally, the situation we are considering is one in which we want a point estimate of a discrete random variable, but the estimate is only to be given if it can be pronounced with confidence (conventionally understood as 95 percent certainty). Think of this as asking for the MAP estimate with an extra condition, namely, that the posterior must be above a certain threshold value $\theta$.

### Analytical results

The standard acceptance rule can be seen as deriving from the Lockean rule, LR. But there are two recent competitors to LR&mdash;to wit, OR and HR&mdash;that might also be seen as giving rise to acceptance rules possibly different from LR. If different indeed, we may want to compare the rules with respect to the criterion of cautiousness mentioned above. But first, are the rules really different?

It turns out that, as acceptance rules in the sense used here&mdash;so rules that are to tell us which of our atomic propositions to accept (equivalently, which world, possibly broadly construed, to deem actual)&mdash;LR and HR amount to the same. Let $\mathfrak{lr}$ be the Lockean acceptance rule and $\mathfrak{hr}$ the Humean acceptance rule. (NB: These rules are not the same as LT and HT from the toy models paper, but rather specific instances of those rules, namely, where the accepted proposition is required to be atomic.) We want to show that $\mathfrak{lr}(\mathcal{H}, \mathcal{E})=\mathfrak{hr}(\mathcal{H}, \mathcal{E})$, for all $\mathcal{H}$, $\mathcal{E}$.

It suffices to note two things. First, we know that HT gives rise to an informative belief set iff
$$
\exists\:\,\text{non-empty}\:\, S\subset W\colon 0 \,<\, \Pr(W\setminus S) \:<\: \frac{1-\theta}{\theta}\cdot\min\bigl(\{p_i\colon w_i\in S\}\bigr).
$$
In the case we are interested in, $S$ is an atomic proposition, which we are now thinking of as one specific member of hypothesis partition $\mathcal{H}$, or alternatively as the (possibly coarse-grained) world $w$ in which that member of $\mathcal{H}$ holds true. Thus, the above criterion simplifies to
$$
\exists i\colon \sum_{j\neq i}\Pr(H_j) \:\,<\:\, \frac{1-\theta}{\theta}\Pr(H_i).
$$
Second,
$$
\mathfrak{lr}(\mathcal{H}, \mathcal{E}) = i \:\:\Leftrightarrow\:\: \Pr_\nolimits\mathcal{E}(H_i) \,>\, \theta \:\:\Leftrightarrow\:\: \Pr_\nolimits\mathcal{E}(H_i) - \theta\Pr_\nolimits\mathcal{E}(H_i) \,>\, \theta - \theta\Pr_\nolimits\mathcal{E}(H_i) \:\:\Leftrightarrow\:\: (1-\theta)\Pr_\nolimits\mathcal{E}(H_i) \,>\, 
\theta\bigl(1 - \Pr_\nolimits\mathcal{E}(H_i)\bigr) \:\:\Leftrightarrow\:\: \frac{1-\theta}{\theta}\Pr_\nolimits\mathcal{E}(H_i)\,>\, 1-\Pr_\nolimits\mathcal{E}(H_i) \:\:\Leftrightarrow\:\: \frac{1-\theta}{\theta}\Pr_\nolimits\mathcal{E}(H_i) \,>\, \sum_{j\neq i}\Pr_\nolimits\mathcal{E}(H_j) \:\:\Leftrightarrow\:\: \mathfrak{hr}(\mathcal{H}, \mathcal{E}) = i.
$$
On the other hand, where $\mathfrak{or}$ is the odds-threshold rule, it can be shown that there are $\mathcal{H}$ and $\mathcal{E}$ such that $\mathfrak{or}(\mathcal{H}, \mathcal{E})\neq\mathfrak{lr}(\mathcal{H}, \mathcal{E})$ (and hence also $\mathfrak{or}(\mathcal{H}, \mathcal{E})\neq\mathfrak{hr}(\mathcal{H}, \mathcal{E})$). In fact, we can show something stronger, to wit, that $\mathfrak{lr}(\mathcal{H}, \mathcal{E})=i \Rightarrow\mathfrak{or}(\mathcal{H}, \mathcal{E})=i$ while $\mathfrak{or}(\mathcal{H}, \mathcal{E})=i \nRightarrow\mathfrak{lr}(\mathcal{H}, \mathcal{E})=i$. Thus, $\mathfrak{or}$ is less cautious than $\mathfrak{lr}$.

To see that $\mathfrak{lr}(\mathcal{H}, \mathcal{E})=i \Rightarrow\mathfrak{or}(\mathcal{H}, \mathcal{E})=i$, suppose that, for particular $\mathcal{H}$, $\mathcal{E}$, $\mathfrak{lr}(\mathcal{H},\mathcal{E})=i$ for some $i\colon 1\leqslant i\leqslant n$. Then $\Pr_\mathcal{E}(H_i)>\theta$ and hence, for all $j\neq i$, $\Pr_\mathcal{E}(H_j)< 1- \theta$. Thus,
$$
\forall j\neq i\colon \Pr_\nolimits\mathcal{E}(H_j)\:\,<\:\,\frac{1-\theta}{\theta}\cdot\theta \:\,<\:\, \frac{1-\theta}{\theta}\cdot\Pr_\nolimits\mathcal{E}(H_i),
$$
while because $(1-\theta)/\theta < 1$, we do have $\Pr_\mathcal{E}(H_i)>\bigl((1-\theta)/\theta\bigr)\cdot\Pr_\mathcal{E}(H_i)$. Hence, $\mathfrak{or}(\mathcal{H},\mathcal{E})=i$.

To see that the converse does not hold, consider the model with $\theta=.95$, $\mathcal{H}=\{H_i\}_{1\leqslant i\leqslant 11}$ a hypothesis partition, and evidence $\mathcal{E}$ such that $\Pr_\mathcal{E}(H_1)=.9$ and $\Pr_\mathcal{E}(H_j)=.01$ for $j\in\{2,\ldots,11\}$. One verifies that $\mathfrak{or}(\mathcal{H},\mathcal{E})=1$ while of course $\mathfrak{lr}(\mathcal{H},\mathcal{E})=\mathtt{void}$.

Hence, we concentrate on $\mathfrak{lr}$ and $\mathfrak{or}$ in the following and ask how to compare in the respect of cautiousness. More exactly, we ask which makes the best trade-off between speed and accuracy.

The setting we consider is one in which we toss a coin with unknown bias 1000 times and must try to determine its bias as the series of tosses unfolds.

### Set parameters, define priors and likelihoods, and generate data

```julia
@everywhere const numb_hyp = 11
@everywhere const numb_toss = 1000
@everywhere const numb_sim = 1000

@everywhere const prior = fill(1/numb_hyp, numb_hyp)
@everywhere const likelihood_heads = linspace(0, 1, numb_hyp)
@everywhere const likelihood_tails = linspace(1, 0, numb_hyp)

datFunc(bias) = rand(Bernoulli(bias), numb_toss)

@everywhere sim_dat = Array{Bool,3}(numb_sim, numb_toss, numb_hyp);

for i in 1:numb_hyp, j in 1:numb_sim
    sim_dat[j, :, i] = datFunc((i - 1) / (numb_hyp - 1))
end
```

### Bayes' rule

```julia
@everywhere function b_upd(probs::Array{Float32,1}, dat::Array{Bool,1}, toss_num::Int)
    if dat[toss_num] == 1
        @. (probs * likelihood_heads) / $dot(probs, likelihood_heads)
    else
        @. (probs * likelihood_tails) / $dot(probs, likelihood_tails)
    end
end
```

### Running the simulations

```julia
bayesRes = SharedArray{Float32,4}(numb_toss + 1, numb_hyp, numb_sim, numb_hyp)

@inbounds for h in 1:numb_hyp, s in 1:numb_sim
    bayesRes[1, :, s, h] = prior
end

@sync @parallel for h in 1:numb_hyp
    @fastmath @inbounds for s in 1:numb_sim, t in 1:numb_toss
        bayesRes[t + 1, :, s, h] = b_upd(bayesRes[t, :, s, h], sim_dat[s, :, h], t)
    end
end
```

### Functions for registering first tosses at which one of the bias hypotheses becomes acceptable, according to the different acceptance rules

```julia
function lrScore(sim::Int, hyp::Int)
    ind = findfirst(x->x==1, sum(bayesRes[:, :, sim, hyp].>.95, 2))
    if ind == 0
        val = 2
    elseif findfirst(x->x>.95, bayesRes[ind, :, sim, hyp]) != hyp
        val = 0
    else
        val = 1
    end    
    return ind, val
end

lrScFnc(hyp) = [ lrScore(i, hyp) for i in 1:numb_sim ]

function firstRightWrongLR(hyp::Int, rightwrong::Int) # for right assignments of high probability: 1; for wrong assignments: 0
    Rs = lrScFnc(hyp)
    out = map(x->x[1], map(collect, Rs))[map(x->x[2], map(collect, Rs)).==rightwrong]
    return out
end
```

```julia
function orScore(sim::Int, hyp::Int)
    f = Array[]
    @inbounds @simd for i in 1:numb_toss
        @fastmath m = maximum(bayesRes[i, :, sim, hyp]) * (1 - .95)/.95
        append!(f, [find(bayesRes[i, :, sim, hyp] .>= m)])
    end
    ff = findfirst(x->length(x) == 1, f)
    if ff == 0 # ff == 0 indicates that there was no unique hypothesis acceptable, only a disjunction of hypotheses
        val = 2
    elseif f[ff][1] == hyp
        val = 1
    else
        val = 0
    end
    return ff, val 
end

orScFnc(hyp) = [ orScore(i, hyp) for i in 1:numb_sim ]

function firstRightWrongOR(hyp::Int, rightwrong::Int) # for right assignments of high probability: 1; for wrong assignments: 0
    Rs = orScFnc(hyp)
    out = map(x->x[1], map(collect, Rs))[map(x->x[2], map(collect, Rs)).==rightwrong]
    return out
end
```

### Speed of convergence

```julia
firstRightLR = [ firstRightWrongLR(i, 1) for i in 2:6 ];
```

```julia
firstRightOR = [ firstRightWrongOR(i, 1) for i in 2:6 ];
```

```julia
@everywhere function bs(x, n=1000)
    bootstrap(x, mean, BasicSampling(n))
end

lrBoot = SharedArray{Float64}(5, 3)
    
@parallel for i in 1:5
    lrBoot[i, :] = [Bootstrap.ci(bs(firstRightLR[i]), BasicConfInt(.95))[1]...]
end

orBoot = SharedArray{Float64}(5, 3)
    
@parallel for i in 1:5
    orBoot[i, :] = [Bootstrap.ci(bs(firstRightOR[i]), BasicConfInt(.95))[1]...]
end
```

```julia
conf_df = convert(DataFrame, vcat(lrBoot, orBoot))
names!(conf_df, [Symbol("$i") for i in ["y", "ymin", "ymax"]])
conf_df[:Bias] = repeat(collect(0.1:0.1:0.5), outer=2)
conf_df[:Rule] = repeat(["LR", "OR"], inner=5);
```

It is clear that OR converges faster than LR, and significantly so for almost all biases (considering the non-overlapping error bars)&mdash;even highly significantly so, as will be seen further down.

#### Accuracy

To compare proportions of mistakes, we use a one-sided *z* test (we already know that OR makes *more* mistakes than LR and now want to know whether it makes *significantly* more mistakes&mdash;hence a one-sided test).

```julia
function propTest(y1::Int64, y2::Int64, n1::Int64, n2::Int64)
    p1 = y1 / n1
    p2 = y2 / n2
    p  = (y1 + y2) / (n1 + n2)
    z = (p1 - p2) / sqrt(p*(1 - p)*((1 / n1) + (1 / n2)))
    return z, cdf(Normal(), z)
end
```

```julia
[ propTest(length(firstWrongLR[i]), length(firstWrongOR[i]), 1000, 1000) for i in 2:6 ]
```

None of the tests comes out significant (even if we do not correct for multiple comparisons): largest $|z|=1.50$, smallest $p = 0.066$. On the other hand, it seems fairer to compare the results aggregated across all possible biases, which is what the following does:

```julia
propTest(length(vcat(firstWrongLR...)), length(vcat(firstWrongOR...)), 11000, 11000)
```

And then the outcome is highly significant.

### A game-theoretic comparison

In the abstract, it is difficult if not impossible to determine how speed and accuracy are to be weighed against each other. In some situations, speed will be more important to us, even if that goes soemwhat at the expense of accuracy; in other situations, the opposite will hold; and in again other situations, they will be equally important. But we can get *some* sense of how much more important accuracy must be than speed if we are to prefer $\mathfrak{lr}$&mdash;the slower but more accurate acceptance rule&mdash;over $\mathfrak{or}$. To that end, we use a variant of a game described on page 434 of [this paper](https://www.researchgate.net/publication/249973626_Inference_to_the_Best_Explanation_Dutch_Books_and_Inaccuracy_Minimisation).

The game involves a magician who has a set of eleven coins in his pocket, where one coin has a perfect bias for heads, another a bias of .1 for heads, a third a bias of.2 for heads, etc. Furthermore, there are two players, one using $\mathfrak{lr}$ as an acceptance rule, the other using $\mathfrak{or}$. The game consists of the magician picking one of the coins in his pocket and starting tossing it. The player are both watching the outcomes of the tosses. A player scores a point by raising her hand and asserting the truth about the bias of the coin. On the other hand, if the player gets the bias wrong, the other player gets the point. If they raise their hands after the same update and are both right about the bias, both receive a point; if they are both wrong, neither receives a point; and if one is right and the other  wrong, then the one who is right receives *two* points. Once a player has asserted a bias hypothesis&mdash;whether the assertion was right or wrong&mdash;the magician puts the coin back into his pocket and picks again randomly a coin out of his pocket. Then the whole procedure starts anew. Each game consists of 100 repetitions of this procedure. The person with most points at the end of the 100 repetitions is the winner.

Who will win?

```julia
@everywhere function lrsc(res::Array{Float32,2}, hyp::Int)
    ind = findfirst(x->x==1, sum(res.>.95, 2))
    if ind == 0
        val = 2
    elseif findfirst(x->x>.95, res[ind, :]) != hyp
        val = 0
    else
        val = 1
    end
    return ind, val
end
```

```julia
@everywhere function orsc(res::Array{Float32,2}, hyp::Int)
    f = Array[]
    @inbounds @simd for i in 1:numb_toss
        @fastmath m = maximum(res[i, :]) * (1 - .95)/.95
        append!(f, [find(res[i, :] .>= m)])
    end
    ff = findfirst(x->length(x) == 1, f)
    if ff == 0
        val = 2
    elseif f[ff][1] == hyp
        val = 1
    else
        val = 0
    end
    return ff, val 
end
```

```julia
function game()   
    r = rand(1:11)
    data = Array{Bool,1}(datFunc((r - 1) / (numb_hyp - 1)))
    updates = Array{Float32}(numb_toss + 1, numb_hyp)
    updates[1, :] = prior
    @fastmath @inbounds for t in 1:numb_toss
        updates[t + 1, :] = b_upd(updates[t, :], data, t)
    end
    locke = 0
    linkelly = 0
    ls1, ls2 = lrsc(updates, r)
    os1, os2 = orsc(updates, r)
    if os1 < ls1 # the case where ls1 < os1 is not possible, in view of the analytical results
        if os2 == 1
            linkelly = 1
        elseif os2 == 0
            locke = 2
        end
    elseif ls1 == os1 && ls2 == os2 == 1
        locke = 1
        linkelly = 1
    end   
    return [locke, linkelly]
end
```

```julia
function run(nSim::Int)
    locke = 0
    linkelly = 0
    k = 0
    while k < nSim
        out = [ game() for i in 1:100 ]
        lo = sum([ out[i][1] for i in 1:100 ])
        lk = sum([ out[i][2] for i in 1:100 ])
        if lo < lk
            linkelly += 1
        elseif lo > lk
            locke += 1
        end
        k += 1
    end
    return locke, linkelly
end
```

We simulate 1000 games, and find ...

```julia
run(1000)
```

 ... that the player with the Lockean strategy *never* wins!

```julia
function game(penalty::Int)   
    r = rand(1:11)
    data = Array{Bool,1}(datFunc((r - 1) / (numb_hyp - 1)))
    updates = Array{Float32}(numb_toss + 1, numb_hyp)
    updates[1, :] = prior
    @fastmath @inbounds for t in 1:numb_toss
        updates[t + 1, :] = b_upd(updates[t, :], data, t)
    end
    locke = 0
    linkelly = 0
    ls1, ls2 = lrsc(updates, r)
    os1, os2 = orsc(updates, r)
    if os1 < ls1 # the case where ls1 < os1 is not possible, in view of the analytical results
        if os2 == 1
            linkelly = 1
        elseif os2 == 0
            locke = penalty
        end
    elseif ls1 == os1 && ls2 == os2 == 1
        locke = 1
        linkelly = 1
    end   
    return [locke, linkelly]
end
```

```julia
function run(nSim::Int, penalty::Int)
    locke = 0
    linkelly = 0
    k = 0
    while k < nSim
        out = [ game(penalty) for i in 1:100 ]
        lo = sum([ out[i][1] for i in 1:100 ])
        lk = sum([ out[i][2] for i in 1:100 ])
        if lo < lk
            linkelly += 1
        elseif lo > lk
            locke += 1
        end
        k += 1
    end
    return locke, linkelly
end
```

```julia
run(1000, 15)
```

```julia
run(1000, 16)
```

It is only when we increase the penalty for making a wrong announcement from 2 to 16 that the Lockean strategy wins about as often as the Odds-threshold strategy.

## Comparison with frequentist statistics?

A frequentist would not update probabilities but would calculate the maximum likelihood estimate, together with a 95 percent confidence interval. We can get something similar to the acceptance rule considered above by declaring a hypothesis acceptable if the CI does not contain any of the other hypotheses. So, for instance, if the CI for the MLE is $[.725, .850]$, then the hypothesis that the bias of the coin equals .8 is acceptable. How does this rule compare to the ones above? 

```julia
function confInt(toss::Int, hyp::Int, sim::Int)
    # uses Agresti-Coull method for calculating 95 percent CIs
    ptilde = ( sum(sim_dat[sim, :, hyp][1:toss]) + .5 * 1.96 ^ 2 ) / (toss + 1.96 ^ 2)
    acUpper = ptilde + 1.96 * sqrt( (ptilde * (1 - ptilde)) / (toss + 1.96 ^ 2) )
    acLower = ptilde - 1.96 * sqrt( (ptilde * (1 - ptilde)) / (toss + 1.96 ^ 2) )

    if ceil(acLower * 10) == floor(acUpper * 10) == hyp - 1
        val = 1
    elseif ceil(acLower * 10) == floor(acUpper * 10) != hyp - 1
        val = 0
    else
        val = 2
    end

    return val
    
end

function findRightWrong(hyp::Int, sim::Int)
    ser = [ confInt(i, hyp, sim) for i in 1:numb_toss ]
    ff = findfirst(x->x!=2, ser)
    res = ser[ff]
    return ff, res
end

function classicRightWrong(hyp::Int, rw::Int) # right: rw = 1; wrong: rw = 0
    frw = [ findRightWrong(hyp, i) for i in 1:numb_sim ]
    firstRW = Int[]
    for i in 1:numb_sim
        if frw[i][2] == rw
            push!(firstRW, frw[i][1])
        else
            nothing
        end
    end
    return firstRW
end
```

```julia
classicMtrx = [ classicRightWrong(i, 1) for i in 2:6 ];
```

```julia
clBoot = SharedArray{Float64}(5, 3)
    
@parallel for i in 1:5
    clBoot[i, :] = [Bootstrap.ci(bs(classicMtrx[i]), BasicConfInt(.95))[1]...]
end
```

We see that the frequentist acceptance rule is much less cautious than either Locke's rule or the Odds-threshold rule. As a result, it is much faster in converging toward the truth, but it also makes many more mistakes along the way. We can again make the comparison more concrete by extending the game used above to include three players.

```julia
@everywhere function cnfI(dat::Array{Bool,1}, toss::Int, hyp::Int)
    ptilde = ( sum(dat[1:toss]) + .5 * 1.96 ^ 2 ) / (toss + 1.96 ^ 2)
    acUpper = ptilde + 1.96 * sqrt( (ptilde * (1 - ptilde)) / (toss + 1.96 ^ 2) )
    acLower = ptilde - 1.96 * sqrt( (ptilde * (1 - ptilde)) / (toss + 1.96 ^ 2) )

    if ceil(acLower * 10) == floor(acUpper * 10) == hyp - 1
        val = 1
    elseif ceil(acLower * 10) == floor(acUpper * 10) != hyp - 1
        val = 0
    else
        val = 2
    end

    return val
    
end

@everywhere function fRW(dat::Array{Bool,1}, hyp::Int)
    ser = [ cnfI(dat, i, hyp) for i in 1:numb_toss ]
    ff = findfirst(x->x!=2, ser)
    res = ser[ff]
    return ff, res
end

@everywhere datFunc(bias) = rand(Bernoulli(bias), numb_toss)

@everywhere function game_ext(penalty::Int)   
    r = rand(1:11)
    data = Array{Bool,1}(datFunc((r - 1) / (numb_hyp - 1)))
    updates = Array{Float32}(numb_toss + 1, numb_hyp)
    updates[1, :] = prior
    @fastmath @inbounds for t in 1:numb_toss
        updates[t + 1, :] = b_upd(updates[t, :], data, t)
    end
        
    ls1, ls2 = lrsc(updates, r)
    os1, os2 = orsc(updates, r)
    cl1, cl2 = fRW(data, r)
    
    locke = 0
    linkelly = 0
    classic = 0

    A = [ls1, os1, cl1]
    A_min = minimum(A)
    fm = find(a->a==A_min, A)
    fp = find(x->x==1, [ls2, os2, cl2][fm])
    fn = find(x->x==0, [ls2, os2, cl2][fm])

    a = [0, 0, 0]
    a[fm[fp]] += 1
    isempty(fn) ? nothing : a[setdiff([1, 2, 3], fm[fn])] += penalty
    locke, linkelly, classic = a
    
    return [locke, linkelly, classic]
end

@everywhere function score_ext(nSim::Int, penalty::Int)
    
    scores = Array{Int,2}(nSim, 3)
    
    @inbounds @fastmath for k in 1:nSim
        out = [ game_ext(penalty) for i in 1:100 ]
        lo = sum([ out[i][1] for i in 1:100 ])
        lk = sum([ out[i][2] for i in 1:100 ])
        cl = sum([ out[i][3] for i in 1:100 ])
        
        fm = find(x->x==maximum([lo, lk, cl]), [lo, lk, cl])
        
        a = [0, 0, 0]
        a[fm] += 1

        scores[k, :] = a
    end
    
    return scores
    
end

@everywhere run_ext(nSim::Int, penalty::Int) = sum(score_ext(nSim, penalty), 1)
```

```julia
steps = collect(2:2:40)
runRes = SharedArray{Int64,2}(length(steps), 3)

@sync @parallel for i in 1:length(steps)
    @fastmath @inbounds runRes[i, :] = [run_ext(1000, steps[i])...]
end
```

Each player receives one point for being the first to announce to true bias hypothesis. The 'penalty' is the number of points the *other* players receive if a player announces a *false* bias hypothesis. We let the penalty vary, from 2 to 40, in increments of 2. It is to be interpreted as indicating how accuracy is weighed against speed (higher penalty means that accuracy is deemed more important relative to speed). The results show that it depends on how we weigh these features against each other whether we are better off with the Lockean acceptance rule, the Odds-threshold acceptance rule, or the frequentist acceptance rule. Specifically, the frequentist rule does best when speed and accuracy are deemed more or less equally important; when we start to value accuracy more, the Odds-threshold rule takes over for a while; when we value accuracy *much* more than speed, the Lockean rule gains the upper hand.
